## Speed up a function: Numba 

```python
import numba as nb

@nb.jit(nopython=True)
def my_add(N): ###
    S = 0
    for i in range(N+1):
        #print("i: ",i,"S before: ",S)
        S += i
        #print("i: ",i,"S_after: ",S)
    return S

i_want = np.arange(10,100+10,1)
### Homework: how to monitor the time used for the loop bellow?
for i in i_want:
    S = my_add(i)
    print(S)
```
